```python
'''Task 1(Import Breast Cancer Dataset)'''
import pandas as pd
dataset=pd.read_csv("C:/Users/BHARATHWAJ T A/Desktop/AIDI Durham/20F/AIDI1002/TASK/dataset.csv")
dataset
```

```python
print(dataset.shape)
```

```python
'''Task 2(Exploratory Analysis of CSV Data)'''
dataset.isnull().sum() #Shows whether there are any null values 
```

```python
datasetLength=len(dataset)

for index in range(datasetLength):
    if "B" in dataset["diagnosis"].iloc[index]:
        dataset.loc[index,"diagnosis"]=1      #Changes from 'B'-Benign to 1
    else:
        dataset.loc[index,"diagnosis"]=2      #Changes from 'M'-Malignant to 2

print(dataset)
```

```python
dataset.describe()
```

```python
'''Task 3 (Replacing missing values to increase accuracy)'''
import numpy as np
dataset=dataset.replace(0, np.NaN)    #Replaces 0 with NaN
```

```python
import missingno as msno
msno.bar(dataset)         #Displays missing values
```

```python
#Above bar graph shows that there are some values missing in few of the columns.
msno.heatmap(dataset)
```

```python
#To prevent missing values, mean of some columns are calculated 
concavityMean=dataset["concavity_mean"].mean()
pointsMean=dataset["points_mean"].mean()
concavitySe=dataset["concavity_se"].mean()
pointsSe=dataset["points_se"].mean()
concavityWorst=dataset["concavity_worst"].mean()
pointsWorst=dataset["points_worst"].mean()
```

```python
dataset["concavity_mean"].fillna(concavityMean, inplace = True)
dataset["points_mean"].fillna(pointsMean, inplace = True)
dataset["concavity_se"].fillna(concavitySe, inplace = True)
dataset["points_se"].fillna(pointsSe, inplace = True)
dataset["concavity_worst"].fillna(concavityWorst, inplace = True)
dataset["points_worst"].fillna(pointsWorst, inplace = True)
#Then the mean values are filled in respective columns  

msno.bar(dataset)  #Below image shows that there are no missing values
```

```python
'''Task 4 Analysing whether there are Outliers'''
Q1 = dataset.quantile(0.25)
Q3 = dataset.quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
print("Lowest limit\n",lower)
print("\n")
print("Highest limit\n",upper)

dataset>upper
```

The above Table shows that most of the values are within the range in this dataset.

```python
'''Task 5 Visual representation of Dataset using Sweetviz library'''
import sweetviz as sv
svReport=sv.analyze(dataset)
svReport.show_html()
```

```python
'''Task 6 Selecting Features for training algorithms'''
y = dataset["diagnosis"]        #diagnosis is output. Therefore used as y variable
y=y.astype('int')
X = dataset.drop(["id","diagnosis"],axis = 1)   #Remaining columns are input 
X.loc[450:500]
```

```python
'''Task 7 This task is used to check the accuracy of each algorithm'''
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
names = ["Decision Tree", "Random Forest", "KNN"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
scores = []

#Decision Tree
DT=DecisionTreeClassifier()
DT.fit(X_train,y_train)
yPredDT = DT.predict(X_test)
scores.append(accuracy_score(y_test, yPredDT))

#Random Forest Classifier
RF=RandomForestClassifier(n_estimators=100)
RF.fit(X_train,y_train)
yPredRF = RF.predict(X_test)
scores.append(accuracy_score(y_test, yPredRF))

#K-Nearest Neighbour
KNN=KNeighborsClassifier(n_neighbors=4)
KNN.fit(X_train,y_train)
yPredKNN = KNN.predict(X_test)
scores.append(accuracy_score(y_test, yPredKNN))

#After calculating Accuracy of each model, it is added inside score variable so that we can display in a form of a table shown below 
algorithmAccuracy = pd.DataFrame({'Algorithm': names, 'Accuracy': scores})
algorithmAccuracy
```

From the above table, it shows that all algorithms are accurate in which the most precise one is Random Forest. The more amount of data with less missing values and Outliers, more will the accuracy for predicting output correctly.
